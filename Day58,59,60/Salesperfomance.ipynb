```python
import matplotlib.pyplot as plt
import numpy as np


months = np.arange(1, 13)


electronics_sales = [45, 50, 48, 55, 60, 65, 62, 70, 75, 80, 85, 90]
clothing_sales = [30, 28, 35, 40, 38, 45, 50, 48, 55, 60, 58, 65]
home_appliances_sales = [20, 22, 25, 28, 30, 33, 35, 37, 40, 42, 45, 50]
books_sales = [15, 17, 18, 16, 20, 22, 24, 23, 25, 27, 29, 30]

fig, axs = plt.subplots(4, 1, figsize=(10, 12), sharex=True)

axs[0].plot(months, electronics_sales, marker='o', color='blue')
axs[0].set_title('Electronics Sales')
axs[0].set_ylabel('Sales (Thousands)')

axs[1].plot(months, clothing_sales, marker='o', color='green')
axs[1].set_title('Clothing Sales')
axs[1].set_ylabel('Sales (Thousands)')

axs[2].plot(months, home_appliances_sales, marker='o', color='orange')
axs[2].set_title('Home Appliances Sales')
axs[2].set_ylabel('Sales (Thousands)')

axs[3].plot(months, books_sales, marker='o', color='purple')
axs[3].set_title('Books Sales')
axs[3].set_xlabel('Month')
axs[3].set_ylabel('Sales (Thousands)')

plt.tight_layout()
plt.show()
```
