```python
import matplotlib.pyplot as plt
import numpy as np

# Months
months = np.arange(1, 13)

# Updated city names
city_names = ['Chennai', 'Madurai', 'Theni', 'Thenkasi']

# Simulated weather data for these cities (same as before, just for example)
temperature = {
    'Chennai': [25, 27, 30, 33, 35, 38, 40, 39, 36, 32, 28, 26],
    'Madurai': [15, 17, 20, 22, 25, 28, 30, 29, 26, 22, 18, 16],
    'Theni': [10, 12, 15, 18, 20, 23, 25, 24, 21, 17, 13, 11],
    'Thenkasi': [30, 32, 35, 37, 40, 43, 45, 44, 41, 36, 33, 31],
}

humidity = {
    'Chennai': [60, 62, 65, 70, 75, 80, 82, 80, 75, 70, 65, 62],
    'Madurai': [50, 52, 55, 57, 60, 62, 65, 64, 60, 58, 55, 53],
    'Theni': [70, 72, 75, 78, 80, 82, 85, 84, 80, 78, 75, 73],
    'Thenkasi': [40, 42, 45, 48, 50, 52, 55, 54, 50, 48, 45, 43],
}

rainfall = {
    'Chennai': [50, 40, 30, 20, 10, 5, 0, 5, 10, 20, 30, 40],
    'Madurai': [60, 55, 50, 45, 40, 35, 30, 35, 40, 45, 50, 55],
    'Theni': [80, 75, 70, 65, 60, 55, 50, 55, 60, 65, 70, 75],
    'Thenkasi': [20, 18, 15, 10, 5, 2, 0, 2, 5, 10, 15, 18],
}

wind_speed = {
    'Chennai': [10, 12, 15, 14, 13, 12, 11, 12, 13, 14, 15, 16],
    'Madurai': [8, 9, 10, 11, 12, 13, 14, 13, 12, 11, 10, 9],
    'Theni': [15, 16, 17, 18, 19, 20, 21, 20, 19, 18, 17, 16],
    'Thenkasi': [5, 6, 7, 8, 9, 10, 11, 10, 9, 8, 7, 6],
}

parameters = ['Temperature (C)', 'Humidity (%)', 'Rainfall (mm)', 'Wind Speed (km/h)']

# Create figure and axes: 4 rows (cities), 4 columns (parameters)
fig, axs = plt.subplots(len(city_names), len(parameters), figsize=(20, 12), sharex=True)

for i, city in enumerate(city_names):
    axs[i, 0].plot(months, temperature[city], marker='o', color='red')
    if i == 0:
        axs[i, 0].set_title(parameters[0])
    axs[i, 0].set_ylabel(city)

    axs[i, 1].plot(months, humidity[city], marker='o', color='blue')
    if i == 0:
        axs[i, 1].set_title(parameters[1])

    axs[i, 2].plot(months, rainfall[city], marker='o', color='green')
    if i == 0:
        axs[i, 2].set_title(parameters[2])

    axs[i, 3].plot(months, wind_speed[city], marker='o', color='purple')
    if i == 0:
        axs[i, 3].set_title(parameters[3])

for ax in axs[-1, :]:
    ax.set_xlabel('Month')

plt.tight_layout()
plt.show()
```
