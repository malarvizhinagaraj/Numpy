```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting
import seaborn as sns




categories = ['Electronics', 'Clothing', 'Home Appliances', 'Books']
category_codes = {cat: i for i, cat in enumerate(categories)}


np.random.seed(0)
num_points = 50
category_data = np.random.choice(categories, num_points)
revenue = np.random.uniform(1000, 10000, num_points)  
customer_count = np.random.randint(10, 200, num_points)


category_numeric = np.array([category_codes[c] for c in category_data])

best_selling_indices = np.argsort(revenue)[-3:]  

months = np.arange(1, 13)
regions = ['North', 'South', 'East', 'West']
monthly_sales = np.random.randint(5000, 20000, size=(len(regions), len(months)))

hours = np.arange(24)
peak_sales = np.random.randint(0, 100, size=(7, 24))  



fig = plt.figure(figsize=(18, 12))


ax1 = fig.add_subplot(2, 2, 1, projection='3d')
scat = ax1.scatter(category_numeric, revenue, customer_count, c=category_numeric, cmap='viridis', s=60, alpha=0.7)
ax1.set_xticks(list(category_codes.values()))
ax1.set_xticklabels(categories)
ax1.set_xlabel('Category')
ax1.set_ylabel('Revenue ($)')
ax1.set_zlabel('Customer Count')
ax1.set_title('3D Scatter: Sales by Category, Revenue & Customer Count')

for i in best_selling_indices:
    ax1.text(category_numeric[i], revenue[i], customer_count[i], f"${int(revenue[i])}", color='red')


ax2 = fig.add_subplot(2, 2, 2, projection='3d')

_x = np.arange(len(months))
_y = np.arange(len(regions))
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
z = np.zeros_like(x)

dx = dy = 0.6
dz = monthly_sales.ravel()

ax2.bar3d(x, y, z, dx, dy, dz, shade=True)
ax2.set_xticks(_x + dx/2)
ax2.set_xticklabels(months)
ax2.set_yticks(_y + dy/2)
ax2.set_yticklabels(regions)
ax2.set_xlabel('Month')
ax2.set_ylabel('Region')
ax2.set_zlabel('Sales ($)')
ax2.set_title('3D Bar Chart: Monthly Sales by Region')


ax3 = fig.add_subplot(2, 1, 2)
sns.heatmap(peak_sales, cmap='YlOrRd', cbar_kws={'label': 'Sales Count'}, ax=ax3)
ax3.set_xlabel('Hour of Day')
ax3.set_ylabel('Day of Week')
ax3.set_title('Heatmap: Peak Sales Hours')
ax3.set_yticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=0)
ax3.set_xticklabels(hours)

plt.tight_layout()
plt.show()
```
