```python
import pandas as pd



df = pd.read_csv('customer_purchases.csv')

print("First 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())


selected_columns = df[["Customer_ID", "Product_Category", "Purchase_Amount"]]
print("\nSelected columns:")
print(selected_columns.head())


electronics_df = df[df["Product_Category"] == "Electronics"]
print("\nCustomers who purchased Electronics:")
print(electronics_df)


high_purchase_df = df[df["Purchase_Amount"] > 500]
print("\nCustomers with Purchase_Amount > $500:")
print(high_purchase_df)


df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
date_filtered = df[df["Purchase_Date"].between("2023-01-01", "2023-12-31")]
print("\nTransactions between 2023-01-01 and 2023-12-31:")
print(date_filtered)


city_filtered = df[df["City"].isin(["New York", "San Francisco"])]
print("\nPurchases in New York or San Francisco:")
print(city_filtered)


print("\nRow using .loc[] (index 2):")
print(df.loc[2])
print("\nRow using .iloc[] (3rd row):")
print(df.iloc[2])

df_indexed = df.set_index("Customer_ID")
print("\nDataFrame with Customer_ID as index:")
print(df_indexed.head())


df_reset = df_indexed.reset_index()
print("\nDataFrame after resetting index:")
print(df_reset.head())

high_value_furniture = df[(df["Product_Category"] == "Furniture") & (df["Purchase_Amount"] > 1000)]
print("\nHigh-value Furniture customers:")
print(high_value_furniture)

items_range = df[df["Items_Purchased"].between(3, 10)]
print("\nCustomers who purchased between 3 and 10 items:")
print(items_range)

filtered_region = df[(df["Purchase_Amount"] > 500) & (df["Region"] == "East")]
print("\nCustomers with Purchase_Amount > $500 and from East region:")
print(filtered_region)
```
