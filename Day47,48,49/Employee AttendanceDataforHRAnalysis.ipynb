```python
import pandas as pd
import numpy as np
data = {
    "Employee_ID": ["E101", "E102", "E103", "E101", "E104", "E105", "E106", "E107", "E108"],
    "Name": ["John Doe", "Jane Smith", "Bob Stone", "John Doe", "Alice Wang", "Ron Miles", "Maya Lee", None, "Jeff King"],
    "Date": ["2023/12/01", "2023-12-01", "01-12-2023", "2023/12/01", "2023.12.01", "2023/12/02", "2023-12-02", "2023-12-02", "12/02/2023"],
    "Check_In": ["09:05", "NULL", "09:12", "09:05", np.nan, "09:00", "08:55", "09:10", "09:30"],
    "Check_Out": ["18:02", "17:58", "18:00", "18:02", np.nan, "NULL", "17:45", "18:10", "18:15"],
    "Status": ["Present", "Present", "Present", "Present", "Absent", "Present", "Present", "Present", "NULL"]
}
df=pd.DataFrame(data)
print(df)

df.replace({"NULL":np.nan,"null":np.nan}, inplace=True)
print(df)

df.drop_duplicates(inplace=True)
print(df)

df['date']=pd.to_datetime(df['Date'], errors='coerce')

df["Check_In"]=pd.to_datetime(df["Check_In"],format='%H:%M')
print(df)
df["Check_Out"]=pd.to_datetime(df["Check_Out"],format='%H:%M')
print(df)

df["Status"].fillna("Unknown", inplace=True)
df["Name"].fillna("Unknown", inplace=True)
df["Check_In"].fillna(pd.NaT, inplace=True)
df["Check_Out"].fillna(pd.NaT, inplace=True)

df.reset_index(drop=True, inplace=True)

print(df)
```
