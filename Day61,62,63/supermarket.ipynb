```python
!pip install numpy
!pip install pandas seaborn matplotlib
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data ={'Category': ['Fruits', 'Vegetables', 'Dairy', 'Beverages', 'Snacks'],
        'Sales': [1500, 1200, 800, 600, 900],
        'Profit': [300, 200, 150, 100, 120] 
        }
df=pd.DataFrame(data)
print(df)
np.random.seed(42)
df['Price']= np.random.randint(5, 20, size=len(df))
df['Revenue'] = df['Sales'] * df['Price']
df['Discount'] = np.random.uniform(0.05, 0.2, size=len(df))
df['Purchase']=np.random.randint(50, 200, size=len(df))



sns.set(style="whitegrid")

plt.Figure(figsize=(12, 6))
sns.barplot(x='Category', y='Revenue', data=df)
plt.title("Average Revenue By Category")
plt.xlabel("Category")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.Figure(figsize=(12, 6))
sns.countplot(x='Category', data=df, palette='viridis') 
plt.title("Count of Products by Category")
plt.xlabel("Category")      
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxenplot(x='Category', y='Price', data=df,palette='coolwarm')
plt.title("Price Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.Figure(figsize=(12, 6))
sns.violinplot(x='Category', y='Discount', data=df, inner='quartile', palette='muted')
plt.title("Discount Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Discount")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.Figure(figsize=(12, 6))
sns.swarmplot(x='Category', y='Sales', data=df, palette='Set2')
plt.title("Sales Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```
