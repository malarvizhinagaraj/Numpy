```python
!pip install seaborn
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)  
categories = ["Data Science", "Machine Learning", "Deep Learning", "AI Ethics"]
difficulty = ["Beginner", "Intermediate", "Advanced"]
data={
    "course":[f"course_{i}" for i in range(1,21)],
     "Category": np.random.choice(categories, 20),
     "Difficulty": np.random.choice(difficulty, 20),
     "Enrollments": np.random.randint(100, 1000, 20),
        "Rating": np.random.uniform(1, 5, 20),
}
df=pd.DataFrame(data)
print(df)


sns.set_style(style="whitegrid")

sns.barplot(x="Category", y="Rating", hue="Difficulty", data=df)

plt.title("Average Course Rating by Category and Difficulty")
plt.xlabel("Course Category")
plt.ylabel("Average Rating")
plt.legend(title="Difficulty Level")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sns.countplot(x="Category",data=df)
plt.title("Number of Courses by Category")
plt.xlabel("Course Category")
plt.ylabel("Number of Courses")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sns.boxplot(x="Category", y="Rating", data=df)
plt.title("Rating Distribution by Course Category")
plt.xlabel("Course Category")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sns.violinplot(x="Category", y="Rating", hue="Difficulty", data=df)
plt.title("Rating Distribution by Course Category and Difficulty")
plt.xlabel("Course Category")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sns.swarmplot(x="Category", y="Enrollments", data=df)
plt.title("Individual Enrollments by Category")
plt.xticks(rotation=30)
plt.show()
```
