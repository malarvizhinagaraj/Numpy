```python
!pip install openpyxl
!conda install openpyxl


import pandas as pd
import sqlite3
import openpyxl


# Read data from different formats
csv_feedback = pd.read_csv("/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/feedback1.csv")  # CSV file

pd.read_excel("feedback2.xlsx", engine='openpyxl')

excel_feedback = pd.read_excel("/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/feedback2.xlsx")  
import pandas as pd

# Correct path (absolute)
df = pd.read_json('/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/feedback3.json')

json_feedback = pd.read_json("/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/feedback3.json") 

import os

file_path = "/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/feedback3.json"

if os.path.exists(file_path):
    json_feedback = pd.read_json(file_path)
    print("JSON loaded successfully")
else:
    print("⚠️ File not found:", file_path)

# Combine into one DataFrame
feedback = pd.concat([csv_feedback, excel_feedback, json_feedback], ignore_index=True)

# Handle missing values (drop rows with no feedback text)
feedback = feedback.dropna(subset=['Feedback'])

# Drop duplicate entries
feedback = feedback.drop_duplicates()

# Optional: Fill missing names or ratings with placeholders
feedback['Name'] = feedback['Name'].fillna('Anonymous')
feedback['Rating'] = feedback['Rating'].fillna(0)

# Connect to SQLite database (or replace ":memory:" with a file path)
conn = sqlite3.connect("feedback.db")

# Store the DataFrame in the SQL database
feedback.to_sql("CustomerFeedback", conn, if_exists='replace', index=False)

feedback.to_csv("cleaned_feedback.csv", index=False)
feedback.to_excel("cleaned_feedback.xlsx", index=False)
feedback.to_json("cleaned_feedback.json", orient='records', lines=True)

chunk_list = []

# Example for large CSV
chunk_list = []

for chunk in pd.read_csv("/Users/malar/Desktop/SQL_MALAR/Numpy/Day50,51,52/large_feedback.csv", chunksize=1000):
    chunk = chunk.dropna(subset=['Feedback'])
    chunk = chunk.drop_duplicates()
    chunk_list.append(chunk)

large_feedback_df = pd.concat(chunk_list, ignore_index=True)
print(large_feedback_df)


# Combine all cleaned chunks
large_feedback_cleaned = pd.concat(chunk_list)

# Save to SQL for further analysis
large_feedback_cleaned.to_sql("LargeFeedback", conn, if_exists='replace', index=False)

print("Memory usage (in bytes):")
print(feedback.memory_usage(deep=True))
```
