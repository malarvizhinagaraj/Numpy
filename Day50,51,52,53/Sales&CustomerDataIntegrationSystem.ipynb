```python
import pandas as pd
# Customer Details
customers = pd.DataFrame({
    'CustomerID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Email': ['alice@example.com', 'bob@example.com', 'charlie@example.com'],
    'Location': ['New York', 'California', 'Texas']
})

# Purchase History
purchases = pd.DataFrame({
    'CustomerID': [1, 2, 2, 3],
    'OrderID': [101, 102, 103, 104],
    'Product': ['Laptop', 'Phone', 'Tablet', 'Laptop'],
    'Amount': [1000, 800, 600, 1000]
})

# Product Details
products = pd.DataFrame({
    'ProductID': [201, 202, 203],
    'Product': ['Laptop', 'Phone', 'Tablet'],
    'Category': ['Electronics', 'Electronics', 'Electronics'],
    'Price': [950, 750, 580]
})

# Merge purchases with customer details on CustomerID
sales_data = pd.merge(purchases, customers, on='CustomerID', how='left')

# Merge with product details on Product
sales_data_full = pd.merge(sales_data, products, on='Product', how='left')

# New customer records
new_customers = pd.DataFrame({
    'CustomerID': [4],
    'Name': ['David'],
    'Email': ['david@example.com'],
    'Location': ['Florida']
})

# Concatenate with original customer data
customers_updated = pd.concat([customers, new_customers], ignore_index=True)

# Drop any exact duplicate columns if they exist
sales_data_full = sales_data_full.loc[:, ~sales_data_full.columns.duplicated()]

# Export to CSV
sales_data_full.to_csv('integrated_sales_customer_data.csv', index=False)

print("ðŸ“Š Final Integrated Sales & Customer Dataset:\n")
print(sales_data_full)
```
