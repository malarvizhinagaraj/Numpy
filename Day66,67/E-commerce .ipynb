```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
n = 200
df = pd.DataFrame({
    'Date': pd.date_range(start='2024-01-01', periods=n, freq='D'),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Home & Kitchen', 'Sports', 'Beauty'], size=n),
    'Price': np.round(np.random.uniform(10, 500, size=n), 2),
    'Sales': np.random.randint(1, 50, size=n),
    'CustomerAge': np.random.randint(18, 65, size=n),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], size=n),
    'Gender': np.random.choice(['Male', 'Female'], size=n)
})
df['TotalRevenue'] = df['Price'] * df['Sales']

sns.set_theme(style="darkgrid", palette="muted")

print(df.head())
print(df.isnull().sum())

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='Category', y='TotalRevenue', estimator=sum, ci=None, palette='viridis')
plt.title("Total Sales by Product Category")
plt.xticks(rotation=30)
plt.savefig("sales_by_category.png", dpi=300)
plt.savefig("sales_by_category.svg", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df['CustomerAge'], bins=10, kde=True, color='skyblue')
plt.title("Customer Age Distribution")
plt.savefig("customer_age_distribution.png", dpi=300)
plt.savefig("customer_age_distribution.svg", dpi=300)
plt.show()

daily_sales = df.groupby('Date')['TotalRevenue'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=daily_sales, x='Date', y='TotalRevenue', marker='o', color='purple', linewidth=2)
plt.title("Sales Trends Over Time")
plt.savefig("sales_trends.png", dpi=300)
plt.savefig("sales_trends.svg", dpi=300)
plt.show()


g = sns.FacetGrid(df, col="Region", col_wrap=2, height=4)
g.map_dataframe(sns.barplot, x="Category", y="Sales", ci=None, estimator=sum, palette="coolwarm")
g.fig.suptitle("Sales by Region", y=1.05)
plt.savefig("sales_by_region.png", dpi=300)
plt.savefig("sales_by_region.svg", dpi=300)
plt.show()


plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Price', y='Sales', hue='Category', palette='husl', s=80)
plt.title("Product Price vs Sales Volume")
plt.savefig("price_vs_sales.png", dpi=300)
plt.savefig("price_vs_sales.svg", dpi=300)
plt.show()


plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Category', y='Price', palette='pastel')
plt.title("Price Distribution by Product Category")
plt.savefig("boxplot_price_by_category.png", dpi=300)
plt.savefig("boxplot_price_by_category.svg", dpi=300)
plt.show()


g = sns.FacetGrid(df, col="Gender", height=4)
g.map_dataframe(sns.histplot, x="Sales", bins=10, kde=True, color="coral")
g.fig.suptitle("Sales Distribution by Gender", y=1.05)
plt.savefig("sales_by_gender.png", dpi=300)
plt.savefig("sales_by_gender.svg", dpi=300)
plt.show()
n
sns.catplot(data=df, x="Category", y="Sales", hue="Region", kind="bar", height=5, aspect=2)
plt.title("Product Sales by Category & Region")
plt.savefig("sales_by_category_region.png", dpi=300)
plt.savefig("sales_by_category_region.svg", dpi=300)
plt.show()


plt.figure(figsize=(8, 6))
corr = df[['Price', 'Sales', 'CustomerAge', 'TotalRevenue']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png", dpi=300)
plt.savefig("correlation_heatmap.svg", dpi=300)
plt.show()
```
